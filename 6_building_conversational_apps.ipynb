# 1. criando um modelo RAG

```python
from langchain.llms import OpenAI
from langchain.chat_models import ChatOpenAI
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import FAISS
from langchain.document_loaders import PyPDFLoader
from langchain.chains import  ConversationalRetrievalChain
from langchain.memory import ConversationBufferMemory

#leitura e preparação dos documentos
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 1500, chunk_overlap = 200)
raw_documents = PyPDFLoader('italy_travel.pdf').load()
documents = text_splitter.split_documents(raw_documents)

#documentos no vector database
db = FAISS.from_documents(documents, embedding=OpenAIEmbeddings())

#memoria
memory = ConversationBufferMemory(
    memory_key= 'chat_history', #entender isso
    return_messages= True
)

#modelo
llm = ChatOpenAI()
```

```python
#lets goooo
qa_chain = ConversationalRetrievalChain.from_llm(llm, retriever=db.as_retriever(), memory=memory, verbose=True)
qa_chain.run({'question': 'give me some review about the Pantheon'})
```

```python
qa_chain.run({'question': 'how are you feeling today?'})
```

# 2. tornando o llm em agentico
## ele acessara os documentos apenas quando decidir necessário

```python
from langchain.agents.agent_toolkits import create_retriever_tool
#preparação das tools
tool = create_retriever_tool(
    db.as_retriever(),
    'italy_travel',
    'searchs and returns documents regarding italy landmarks, restaurant and hotels'
)
tools = [tool]

#setup da memoria
memory = ConversationBufferMemory(
    memory_key='chat_history',
    return_messages=True)

from langchain.agents.agent_toolkits import create_conversational_retrieval_agent
from langchain.chat_models import ChatOpenAI
llm = ChatOpenAI()
agent_executor = create_conversational_retrieval_agent(llm, tools, memory_key='chat_history',verbose=True)
```

```python
agent_executor({'input':'tell me something about the pantheon'})
```

```python
output = agent_executor({'input':' what can i do in india for 3 days?'})
```

# 3. Inluindo tools de acesso a internet no nosso agente

```python
from langchain import SerpAPIWrapper
from langchain.agents import AgentType, initialize_agent
from langchain.llms import OpenAI
from langchain.tools import BaseTool, StructuredTool, Tool, tool

load_dotenv()
os.environ['SERPAPI_API_KEY']

search = SerpAPIWrapper()

tools = [
    Tool.from_function(
        func = search.run,
        name = 'search',
        description='searchs the internet for the information regarding current information you do not have access through the documents'
    ),
    create_retriever_tool(
        db.as_retriever(),
        name = 'italy_travel',
        description= 'searchs and returns documents regarding italy landmarks, restaurant and hotels'
    )]

agent_executor = create_conversational_retrieval_agent(llm, tools, memory_key='chat_history',verbose=True)
```

```python
agent_executor({'input':'what can i visit in india in 3 days?'})
```

```python
agent_executor({'input':'what is the weather in delhi?'})
```

```python
agent_executor({'input':'what is the top 3 atractions to visit in italy?'})
```
